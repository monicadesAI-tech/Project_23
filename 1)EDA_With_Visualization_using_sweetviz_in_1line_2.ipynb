```python
import pandas as pd
import sweetviz
```

```python
# Download (train.csv and test.csv) dataset from this below given link and store it same working folder where code is.

# https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
train.tail()
```

```python
# Analyzing single dataframe (its optional target feature)
```

```python
my_report = sweetviz.analyze([train,"Train"],target_feat = 'SalePrice')
```

```python
my_report.show_html('House_output_report.html')
```

```python
# Comparing two dataframes (Train Vs Test)
```

```python
my_report1 = sweetviz.compare([train,"Train"],[test,"Test"], "SalePrice")
```

```python
my_report1.show_html('House_output_report2.html')
```
